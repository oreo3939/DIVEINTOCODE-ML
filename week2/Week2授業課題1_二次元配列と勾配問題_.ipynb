```
import numpy as np
import matplotlib.pyplot as plt
```

##【問題1】1次関数
x が-50から50の範囲で次の式を考えます。

$y = \frac{1}{2}x + 1$

-50から50まで0.1間隔のxを表すndarrayと、各要素に対するyを表すndarrayを作成してください。

```
x_ndarray = np.arange(-50, 50.1, 0.1)#50.1にしないと50まで表示されない
y_ndarray = 1/2 * x_ndarray + 1
```

##【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。0列目にx、1列目にyが来るようにします。

```
x = np.reshape(x_ndarray,([1001,1]))
y = np.reshape(y_ndarray,([1001,1]))
xy = np.concatenate([x,y],axis=1)

print(xy.shape)
```

##【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。

image.png


各変化量は以下の図のように隣同士の要素の差から近似的に求めます。


image.png

《注意》


厳密に各点における勾配を求めるには極限（微分）を利用する必要があります
勾配はx座標の間隔を使うので、xの要素数に対して一つ少なくなります

```
dx = x[1:] - x[:-1]
dy = y[1:] - y[:-1]
dxy = dy / dx


print(dxy)
```

##【問題4】グラフを描く
1次関数のグラフをMatplotlibを用いて描いてください。また、それとは別に横軸をx、縦軸を勾配としたグラフも描いてください。



```
plt.plot(x_ndarray,y_ndarray)
plt.figure()
plt.plot(dx,dxy)
plt.show()
```

##【問題5】Pythonの関数化
問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。


そして、以下の3式の勾配を0.1間隔で求めてndarrayで出力してください。

  　　　　　　$y = x^{2} \quad (-50\leq x \leq 50)\\
y = 2x^{2}+2^{x} \quad (-50\leq x \leq 50)\\
y = sin(x^{\frac{1}{2}}) \quad (0\leq x \leq 50)$


また、問題4と同様にグラフを描いてください。


ndarrayに対して三角関数のsinの計算を行うには np.sin()を用います。

```
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):

  x = np.arange(x_range[0],x_range[1],x_range[2])
  y = function(x)
  xy = np.concatenate([x[:, np.newaxis], y[:, np.newaxis]], axis=1)
  
  xd = x[1:] - x[:-1]
  yd = y[1:] - y[:-1]
  xyd = yd / xd
  return xy, xyd

def function1(x):
  array_y = x**2
  return array_y
def function2(x):
  array_y = 2*(x**2)+2**x
  return array_y
def function3(x):
  array_y = np.sin(x**(1/2))
  return array_y

```

```
array_xy1, gradient1 = compute_gradient(function1)
print(array_xy1)
```

```
plt.plot(array_xy1[:,0],array_xy1[:,1])
plt.figure()
plt.plot(array_xy1[:-1,0],gradient1)#
```

```
array_xy2, gradient2 = compute_gradient(function2)
print(array_xy2)
```

```
plt.plot(array_xy2[:,0],array_xy2[:,1])
plt.figure()
plt.plot(array_xy2[:-1,0],gradient2)
```

```
array_xy3, gradient3 = compute_gradient(function3,x_range=(0,50.1,0.1))
print(array_xy3)
```

```
plt.plot(array_xy3[:,0],array_xy3[:,1])
plt.figure()
plt.plot(array_xy3[:-1,0],gradient3)
```

##【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。


今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。


また、最小のyをとる前後の勾配を表示してください。

```
y_min1 = np.min(array_xy1[:,1])
print(y_min1)
y_argmin1 = np.argmin(array_xy1[:,1])
print(y_argmin1)

print("最小のyの前の勾配{}".format(gradient1[y_argmin1-1]))
print("最小のyの後の勾配{}".format(gradient1[y_argmin1+1]))
```

```
y_min2 = np.min(array_xy2[:,1])
print(y_min2)
y_argmin2 = np.argmin(array_xy2[:,1])
print(y_argmin2)

print("最小のyの前の勾配{}".format(gradient2[y_argmin2-1]))
print("最小のyの後の勾配{}".format(gradient2[y_argmin2+1]))
```

```
y_min3 = np.min(array_xy3[:,1])
print(y_min3)
y_argmin3 = np.argmin(array_xy3[:,1])
print(y_argmin3)

print("最小のyの前の勾配{}".format(gradient3[y_argmin3-1]))
print("最小のyの後の勾配{}".format(gradient3[y_argmin3+1]))
```

##【問題2】SSEを求める関数の作成

クラスタ内誤差平方和（SSE, Sum of Squared Errors）を計算する関数を作成してください。


K-meansはこのSSEを最小化する 
r
n
k
 と 
μ
k
 を求めることが目的となります。複数個の初期値で計算したクラスタリング結果から、どれを最終的に採用するかを決める際にこのSSEを求める関数を使用します。
 
 $$SSE = \sum_{n=1}^N \sum_{k=1}^K r_{nk} \|X_n - \mu_k\|^2$$
 
 n
 : データ点のインデックス


k
 : クラスタのインデックス


X
n
 : 
n
番目のデータ点


μ
k
 : 
k
番目の中心点


r
n
k
 : データ点
X
n
がクラスタ
k
に所属していたら1、そうでなければ0



```python
def sse_c(X, label, center):
    clu = len(center)
    sse = 0
    for k in range(clu):
        norm = X[label == k] - center[k]
        sse += np.linalg.norm(norm)**2
    return sse
```

```python
class ScratchKMeans():

    def __init__(self, n_clusters, n_init, max_iter, tol):
        self.n_clusters = n_clusters #クラスタ数
        self.n_init = n_init #中心点の初期値を何回計算するか
        self.max_iter = max_iter #最大繰り返し数
        self.tol = tol #終了基準
        
        self.best_sse = None #ベストスコア
        self.best_center = None #ベストセンター
        
    def fit(self, X):
        """
        クラスタリングを計算
        
        """
        #サンプル数
        self.num = len(X)
        
        # 初期値計算
        for i in range(self.n_init):
            
            # 中心点作成
            center_idx   = np.random.randint(0, self.num, self.n_clusters)
            self.center = np.array([X[i] for i in center_idx])
            
            # ラベルを初期化
            self.labels = np.random.randint(0, self.n_clusters, self.num)

            # 中心点の更新
            for _ in range(self.max_iter):
                new_center = np.array([X[self.labels == k, :].mean(axis = 0)
                                       for k in range(self.n_clusters)])
                
                # 空のクラスタがあった場合は中心点をランダムな点に割り当てなおす
                for n in range(self.n_clusters):
                    if not np.any(new_center == n):
                        self.center[n] = X[np.random.choice(X.shape[0], 1), :]

                if np.allclose(self.center, new_center, atol= self.tol):
                    break
                
                self.center = new_center
                
                # 所属クラスタの更新
                # 一番近い中心点のクラスタを所属クラスタに更新
                dist = np.array([np.linalg.norm(X- self.center[k], axis=1)
                                 for k in range(self.n_clusters)])
            
               # 最小値のインデックスを取得
                self.labels = np.argmin(dist, axis=0)
            
            # sseを計算しベストスコアであれば更新
            new_sse_score = self._sse(X)
            
            if self.best_sse is None or self.best_sse > new_sse_score:
                self.best_sse = new_sse_score
                self.best_center = self.center
            
            
    def predict(self, X):
        """
        入力されたデータがどのクラスタに属するかを計算
        """
       # 所属クラスタをbest_centerで更新
        dist = np.array([np.linalg.norm(X - self.best_center[k], axis=1)
                         for k in range(self.n_clusters)])
        
        # 最小値のインデックスを取得
        pred = np.argmin(dist, axis=0)
        
        return pred
    
    def _sse(self, X):
        """
        SSEを計算する
        """
        sse = 0
        for k in range(self.n_clusters):
            diff = X[self.labels == k] - self.center[k]
            sse += np.linalg.norm(diff)**2

        return sse
    
    def _update(self, X):
        """
        中心点を更新する
        """
        
        new_center = np.array([X[self.labels == k, :].mean(axis = 0) 
                               for k in range(self.n_clusters)])
        
        dist = np.array([np.linalg.norm(X- self.center[k], axis=1)
                         for k in range(self.n_clusters)])
        
        labels = np.argmin(dist, axis=0)
        return labels
```

```python
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples

def silhouette(data,y_km,k):
    km = KMeans(n_clusters=k, 
                init='k-means++', 
                n_init=10, 
                max_iter=350,
                tol=1e-04,
                random_state=0)
    y_km = km.fit_predict(data_pca)

    cluster_labels = np.unique(y_km)
    n_clusters = cluster_labels.shape[0]
    silhouette_vals = silhouette_samples(data_pca, y_km, metric='euclidean')
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
                 edgecolor='none', color=color)

        yticks.append((y_ax_lower + y_ax_upper) / 2.)
        y_ax_lower += len(c_silhouette_vals)

    silhouette_avg = np.mean(silhouette_vals)
    plt.axvline(silhouette_avg, color="red", linestyle="--")

    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.tight_layout()
    plt.show()
```
